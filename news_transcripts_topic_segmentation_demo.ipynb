## Common funcs

```python
def recover_segments(sentences, seg_bounds):
    segments = []
    last_bound = 0

    # append segments from 0 to n-1
    for sent_idx, (is_bound) in enumerate(seg_bounds):
        if is_bound:
            segment = sentences[last_bound:sent_idx+1]
            segments.append(segment)
            last_bound = sent_idx+1

    # append last segment
    segment = sentences[last_bound:]
    segments.append(segment)

    return segments
```

```python
from pprint import pprint

def print_segments(sections, headings=None, scores=None):
    print(f"SECTION COUNT: {len(sections)}")
    for i, section in enumerate(sections):
        print(f'\n-- SECTION {i+1} START --')
        if headings is not None:
            print(f'-- HEADING: {headings[i]}')
        pprint(section)
        print('-- SECTION END --')
```

## Text segmentation

### Load dataloader

```python
transcript_df = pd.read_csv(transcript_path)
```

```python
print('Total utterances:', transcript_df.shape[0])
transcript_df.head()
```

```python
from models.EncoderDataset import SentenceDataset
import nltk

# use utterances as a unit
sentences = transcript_df['text'].tolist()
sentences = []

# break utterances into sentences and
# use sentences as a unit
for utterance in transcript_df['text']:
    new_sentences = nltk.sent_tokenize(utterance)
    sentences.extend(new_sentences)

tag_to_ix = {0:0, 1:1, '<START>':2, '<STOP>':3} # обязательно
dataset = SentenceDataset([sentences], tag_to_ix, encoder="cointegrated/rubert-tiny2", infer=True)
```

```python
dataset[0]['embeddings'].shape
```

```python
from torch.utils.data import DataLoader
import torch

dataloader = DataLoader(dataset, batch_size=8, collate_fn=dataset.collater)
batch = next(iter(dataloader))
batch
```
